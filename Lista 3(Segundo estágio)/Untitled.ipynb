```python
import sympy as sp
import numpy as np
from sympy import *
from utils import symdisp, symplot
```

```python
m = input('Digite os últimos 3 dígitos de sua Matricula:')
x = round(int(m)/100)+1
y = round((int(m)%100)/10)+1
z = round(int(m)%10)+1
```

```python
t = sp.symbols('t', real=True)
C = 500*10**-6
L = 100*10**-3
v_C_inicial = -x
i_L_inicial = 0.1*y
i = 500*sp.exp(-z*t)*cos(20*t)*(10**-3)
v1 = -2*sp.exp(-y*t)
v2 = 5*t*sp.exp(-x*t)
symdisp('C = ',C*10**6,'uF')
symdisp('L = ',L*10**3,'mH')
symdisp('v_C (0) = ',v_C_inicial,'V')
symdisp('i_L (0) = ',i_L_inicial,'A')
symdisp('i(t) = i_C(t) = ',i*10**3,'mA')
symdisp('v_1(t) = ',v1,'V')
symdisp('v_2(t) = ',v2,'V')
```

# Letra A


```python
#Tensão no Capacitor
v_C_t = (C**-1)*sp.integrate(i,(t,0,t))+v_C_inicial
symdisp('v_C(t) = ',sp.simplify(v_C_t),'V')
#Tensão no Indutor
i_L_t = (sp.diff(i,t) - i + ((v1+v2)/L))
v_L_t = L*i_L_t
symdisp('v_L(t) = ',sp.simplify(v_L_t),'V')
```

```python
#tensões
intervalo = np.arange(0,1.5,0.005)
symplot(t,[v_C_t,v_L_t],intervalo,['Tensão no capacitor','Tensão no indutor'])
```

Verificamos que NÃO É POSSÍVEL VARIAÇÃO ABRUPTA DE TENSÃO NO CAPACITOR

```python
#correntes
intervalo = np.arange(0,1.5,0.005)
symplot(t,[i,i_L_t],intervalo,['Corrente no Capacitor', 'Corrente no Indutor'])
```

Verificação de que NÃO HÁ VARIAÇÃO ABRUPTA DE CORRENTE NO INDUTOR

# Letra B


```python
#Potência na fonte de corrente
v_i = v_L_t - v1 - v_C_t
p_i = v_i*i
symdisp('p_{i(t)}(t) = ',sp.simplify(p_i),'W')
```
