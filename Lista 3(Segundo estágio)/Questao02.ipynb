```python
import sympy as sp
import numpy as np
from sympy import *
from utils import symdisp, symplot
```

```python
m =[8,1,4]#>>>>>>> OS 3 ULTIMO DÍGITOS DA MATRÍCULA <<<<<<<<<<<<<
x = m[0]+1
y = m[1]+1
z = m[2]+1
```

```python
t = sp.symbols('t', real=True)
C = 500*10**-6
L = 100*10**-3
v_C_inicial = -x
i_L_inicial = 0.1*y
i = 500*sp.exp(-z*t)*cos(20*t)*(10**-3)
v1 = -2*sp.exp(-y*t)
v2 = 5*t*sp.exp(-x*t)
symdisp('C = ',C*10**6,'uF')
symdisp('L = ',L*10**3,'mH')
symdisp('v_C (0) = ',v_C_inicial,'V')
symdisp('i_L (0) = ',i_L_inicial,'A')
symdisp('i(t) = i_C(t) = ',i*10**3,'mA')
symdisp('v_1(t) = ',v1,'V')
symdisp('v_2(t) = ',v2,'V')
```

# Letra A


```python
#Tensão no Capacitor
v_C_t = (C**-1)*sp.integrate(i,(t,0,t))+v_C_inicial
symdisp('v_C(t) = ',v_C_t,'V')

#Tensão no Indutor
diff_i_beta = ((v1+v2)/L) + sp.diff(i,t)
diff_i_L_t  =  diff_i_beta - sp.diff(i,t)
v_L_t = L*diff_i_L_t
symdisp('v_L(t) = ',v_L_t,'V')
```

```python
#tensões
intervalo = np.arange(0,3,0.005)
symplot(t,[v_C_t,v_L_t],intervalo,['Tensão no capacitor','Tensão no indutor'])
```

Verificamos que NÃO É POSSÍVEL VARIAÇÃO ABRUPTA DE TENSÃO NO CAPACITOR

```python
#corrente no Capacitor e no Indutor
i_L_t = sp.integrate(v_L_t,(t,0,t))+i_L_inicial
intervalo = np.arange(0,3,0.005)
symplot(t,[i,i_L_t],intervalo,['Corrente no Capacitor','Corrente no Indutor'])
```

Verificação de que NÃO HÁ VARIAÇÃO ABRUPTA DE CORRENTE NO INDUTOR

# Letra B


```python
#Potência na fonte de corrente
v_i = v_L_t - v1 - v_C_t
p_i = v_i*i
symdisp('P_{i(t)}(t) = ',p_i,'W')
intervalo = np.arange(0,1,0.005)
symplot(t,p_i,intervalo,'Potência na fonte de corrente')
```

# Letra C

```python
#Corrente em V1 é i_L_t
#Corrente em V2 é i_L_t + i_C_t, i_C_t == i
#Logo a potência nas fontes serão:
p_v1 = v1*i_L_t
symdisp('P_{v_1(t)}(t) =',p_v1,'W')

p_v2 = v2*(i+i_L_t)
symdisp('P_{v_2(t)}(t) =',p_v2,'W')


intervalo = np.arange(0,3,0.005)
symplot(t,[p_v1,p_v2],intervalo,['Potência na fonte de Tensão V1','Potência na fonte de Tensão V2'])
```

# Letra D

```python
# C = q/v_C_t
q = C*v_C_t
q_oo = sp.limit(q,t,oo)
print('Carga total armazenada no capacitor em t -> oo')
symdisp('lim_{t->+oo} q(t) =',round(q_oo*10**3,3),'mC')
```

# Letra E


```python
W_C = (1/2)*C*(v_C_t**2)
W_C_oo = sp.limit(W_C,t,oo)
print('Energia acumulada no capacitor em t -> oo')
symdisp('lim_{t->+oo} W_C(t) =',round(W_C_oo*10**3,3),'mJ')

W_L = (1/2)*L*(i_L_t**2)
W_L_oo = sp.limit(W_L,t,oo)
print('Energia acumulada no Indutor em t -> oo')
symdisp('lim_{t->+oo} W_L(t) =',round(W_L_oo,3),'J')
```

# Letra F

```python
print('Tensão e Corrente no Capacitor em t-> oo')
v_C_t_oo = sp.limit(v_C_t,t,oo)
i_oo = sp.limit(i,t,oo)
symdisp('lim_{t->+oo} V_C(t) =',round(v_C_t_oo,3),'V')
symdisp('lim_{t->+oo} i_C(t) =',round(i_oo,3),'A')

print('\nTensão e corrente no Indutor em t-> oo')
v_L_t_oo = sp.limit(v_L_t,t,oo)
i_L_t_oo = sp.limit(i_L_t,t,oo)
symdisp('lim_{t->+oo} V_L(t) =',round(v_L_t_oo,3),'V')
symdisp('lim_{t->+oo} i_L(t) =',round(i_L_t_oo,3),'A')
```
