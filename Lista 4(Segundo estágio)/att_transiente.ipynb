```python
import sympy as sp
import numpy as np
from sympy import *
from utils import symdisp, symplot, round_expr
```

$Transiente$: $Chaveamento$ $1$

```python
τ = (300*10**-3)/((5*9*sp.sqrt(2))/(5+ 9*sp.sqrt(2)))
t = sp.symbols('t', real =True)
transiente1 = sp.exp(-t*τ**-1)
symdisp('Ts_1 =',round_expr(transiente1,4))
```

$Transiente$: $Chaveamento$ $2$

```python
τ = (300*10**-3)/((5*2*sp.sqrt(3))/(5+ 2*sp.sqrt(3)))
t = sp.symbols('t', real =True)
transiente2 = sp.exp(-t*τ**-1)
symdisp('Ts_2 =',round_expr(transiente2,4))
```

```python
intervalo  = np.linspace(0,1, 1000)
symplot(t,[transiente1,transiente2],intervalo,funLabel = ['$Transiente$ $1$','$Transiente$ $2$'])
```
